```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Mobility
This notebook is meant to be used in a directory containing all output files resulting from friction simulations.

Firstly, important constants are defined, starting from the value $g$, and some useful functions are defined.
These functions are used to write a "forces.txt" file containing the forces used in simulations (achieved by reading the name of files present in the directory) and the "cut" timestep, i.e. the timestep which separates the transient phase from the steady state. The default value of cut timestep when creating "forces.txt" or updating it is set to 1000.

After plotting the data, the user should decide where the transient phase ends in each configuration (the "cut" is shown as a vertical dashed red line in each plot) and manually update the cut timestep in "forces.txt" accordingly.

```python
g=0.07   # --- define g ---

k=2e-7
a_coll=5.8
V0=g*k*a_coll**2
timestep=50   #microseconds

F1s=4.279/5.4*V0

damp=5000
mass=31
```

```python
import glob
import re
import os

def extract_forces(folder_path):
    # Find all matching files
    file_paths = glob.glob(f"{folder_path}/pos_vel_*.dat")
    
    numbers = []
    for file_path in file_paths:
        match = re.search(r'pos_vel_([0-9]*\.?[0-9]+)\.dat$', file_path)
        if match:
            numbers.append(float(match.group(1)))  # Convert extracted string to float

    return np.array(sorted(numbers))  # Return sorted array
```

```python
# writes forces.txt file with all forces whose .dat file is present,
# setting cut_timestep=1000 as default
def create_forces_file():
    # Read files contained in directory and extract values of F/F_1s used:
    frac= extract_forces(".")
    force=frac*F1s   # the actual forces in simulation units
    
    row = "{:<7} {:>25} {:>10}"
    header_row = "{:<7} {:>25} {:>10}"
    headers = "# F/F_1s F cut_timestep".split()
    
    with open('forces.txt', "w") as outFile:
        #outFile.write("# F/F_1s   F   cut_timestep\n")
        outFile.write(header_row.format("# F/F_1s", "F     ", "cut_timestep\n"))
        for frac_, force_ in zip(frac, force):
            outFile.write(row.format(frac_, force_, str(1000)+"\n"))   # assumes 1000 as default cut value

# updates forces.txt if any .dat files have been added or removed
def update_forces_file():
    frac_new= extract_forces(".")
    force=frac_new*F1s
    frac_old, timesteps= np.loadtxt("forces.txt", unpack=True, usecols=(0,2))
    
    # Create a mapping from frac_old to timesteps
    frac_to_time = dict(zip(frac_old, timesteps))
    
    # Generate the new timesteps array
    new_timesteps = np.array([frac_to_time.get(val, 1000) for val in frac_new])
    
    row = "{:<7} {:>25} {:>10}"
    header_row = "{:<7} {:>25} {:>10}"
    headers = "# F/F_1s F cut_timestep".split()
        
    with open('forces.txt', "w") as outFile:
        #outFile.write("# F/F_1s   F   cut_timestep\n")
        outFile.write(header_row.format("# F/F_1s", "F     ", "cut_timestep\n"))
        for frac_new_, force_, new_timesteps_ in zip(frac_new, force, new_timesteps):
            #print(frac_new_)
            outFile.write(row.format(frac_new_, force_, str(int(new_timesteps_))+"\n"))   # assumes 1000 as default cut

# writes new forces.txt if not present, updates it if already present
def set_forces_file():
    if os.path.isfile("forces.txt"):
        update_forces_file()
    else:
        create_forces_file()
```

## X direction

```python
set_forces_file()

# frac is F/F_1s ;   force is the actual force F in simulation units
frac, force, cut_ts=np.loadtxt("forces.txt", unpack=True, usecols=(0,1,2))

velo=np.zeros(frac.size)  # array containing steady-state velocities

plt.figure(figsize=(10,frac.size*5))

for i in range(frac.size):
    ts, xcm, vcm=np.loadtxt('pos_vel_'+str(frac[i])+'.dat', unpack=True, usecols=(0,2,3))   # pos_vel_0.01.dat
    step=ts[1]-ts[0]
    
    plt.subplot(frac.size+1,2,2*i+1)
    plt.plot(ts, xcm)
    plt.axvline(cut_ts[i], linestyle='dashed', c='r', linewidth=1)
    plt.title('F = '+str(frac[i])+' F1s')
    plt.xlabel('timestep', loc='left')
    plt.ylabel(r'$x_{cm}$')
    plt.grid()

    plt.subplot(frac.size+1,2,2*i+2)
    plt.plot(ts, vcm)
    plt.axvline(cut_ts[i], linestyle='dashed', c='r', linewidth=1)
    plt.title('F = '+str(frac[i])+' F1s')
    plt.xlabel('timestep', loc='left')
    plt.ylabel(r'$v_{x,cm}$')
    plt.grid()

    cut_index=int(cut_ts[i]/step)
    velo[i]=(xcm[-1]-xcm[cut_index])/((ts[-1]-ts[cut_index])*timestep)   # steady-state velocity calculated as dx/dt
    
plt.show()
```

```python
freemob=damp/mass

mobility=velo/force/freemob   # normalized mobility

plt.figure(figsize=(8,5))

plt.plot(frac, mobility, marker='s',
         markersize=4, linewidth=1, label='g='+str(g), c='b')
#plt.xlim(left=0)
plt.xlim(left=0, right=0.7)
plt.ylim(bottom=0, top=1)
plt.legend(loc='lower right')
plt.xlabel(r'$F/F_{1s}$', fontsize=11)
plt.ylabel(r'$\mu/\mu_{free}$', fontsize=12)
plt.show()
```

```python
# print mobility values to output file
with open('mobility_'+str(g)+'.out', "w") as f:
    for i in range(frac.size):
        f.write(str(str(frac[i]))+' '+ str(force[i])+' '+str( mobility[i])+'\n')
f.close()
```

```python
def fit_function(x, A, F_s, p):
    return A*pow(x-F_s,p)

from scipy.optimize import curve_fit
```

```python
# Define some starting parameters ("guess")
A_est=1.2
F_s_est=0.028
p_est=0.5
# Test these guess parameteres if you want (un-comment the plt.plot(x_try, y_try) line to plot this)
x_try=np.linspace(F_s_est, F_s_est+0.11, 100)
y_try=fit_function(x_try, A_est, F_s_est, p_est)

initial_guess = [A_est, F_s_est, p_est]  # Initial guess for values A, F_s, and p (allows the algorithm to converge)

# Define range of F values to perform fit on
fit_start=3
fit_stop=13
print("Performing fit on forces:")
print(frac[fit_start:fit_stop])
params, covariance = curve_fit(fit_function, frac[fit_start:fit_stop], mobility[fit_start:fit_stop], p0=initial_guess)

# Extract fitted parameters
A_fit, F_s_fit, p_fit = params
print("Fitted parameters:")
print("F_s : " + str(F_s_fit))
print("p : " + str(p_fit))

x_fit=np.linspace(F_s_fit, frac[-1], 100)
y_fit=fit_function(x_fit, A_fit, F_s_fit, p_fit)

plt.figure(figsize=(8,5))

plt.plot(frac, mobility, marker='s', c='b', label=r'$g=0.07$',
         linewidth=1, markersize=4)
#plt.plot(x_try, y_try)
plt.plot(x_fit, y_fit, label='Fit', linestyle='dashed', c='r', linewidth=1.2)
plt.xlim(left=0, right=0.7)
plt.ylim(bottom=0, top=1)
plt.legend(loc='lower right', frameon=False)
plt.xlabel(r'$F/F_{1s}$', fontsize=11)
plt.ylabel(r'$\mu/\mu_{free}$', fontsize=12)
plt.savefig("fit_0.07.pdf",bbox_inches='tight')   # save figure as pdf
plt.show()
```

## Y direction

```python
set_forces_file()

# frac is F/F_1s ;   force is the actual force F in simulation units
frac, force, cut_ts=np.loadtxt("forces.txt", unpack=True, usecols=(0,1,2))

veloy=np.zeros(frac.size)  # array containing steady-state velocities

plt.figure(figsize=(10,frac.size*5))

for i in range(frac.size):
    ts, ycm, vy=np.loadtxt('pos_vel_'+str(frac[i])+'.dat', unpack=True, usecols=(0,4,5))   # pos_vel_0.01.dat
    step=ts[1]-ts[0]
    
    plt.subplot(frac.size+1,2,2*i+1)
    plt.plot(ts, ycm)
    plt.axvline(cut_ts[i], linestyle='dashed', c='r', linewidth=1)
    plt.title('F = '+str(frac[i])+' F1s')
    plt.xlabel('timestep', loc='left')
    plt.ylabel(r'$y_{cm}$')
    plt.grid()

    plt.subplot(frac.size+1,2,2*i+2)
    plt.plot(ts, vy)
    plt.axvline(cut_ts[i], linestyle='dashed', c='r', linewidth=1)
    plt.title('F = '+str(frac[i])+' F1s')
    plt.xlabel('timestep', loc='left')
    plt.ylabel(r'$v_{y,cm}$')
    plt.grid()

    cut_index=int(cut_ts[i]/step)
    veloy[i]=(ycm[-1]-ycm[cut_index])/((ts[-1]-ts[cut_index])*timestep)   # steady-state velocity calculated as dy/dt
    
plt.show()
```

```python
#freemob=damp/mass

mobility_y=veloy/force/freemob

plt.figure(figsize=(8,5))

plt.plot(frac, mobility_y, marker='s',
         markersize=4, linewidth=1, label='g='+str(g), c='b')
#plt.xlim(left=0)
plt.xlim(left=0, right=0.7)
plt.ylim(bottom=0, top=1)
plt.legend(loc='upper right', frameon=False)
plt.xlabel(r'$F/F_{1s}$', fontsize=11)
plt.ylabel(r'$\mu_y/\mu_{free}$', fontsize=12)
plt.show()
```

```python
start=2   # compute angle for unpinned configurations only (otherwise v_y/v_x diverges)
angle=np.arctan(mobility_y[start:]/mobility[start:])*180/math.pi

plt.figure(figsize=(8,5))

plt.plot(frac[start:], angle, marker='s',
         markersize=4, linewidth=1, label='g='+str(g), c='r')
#plt.xlim(left=0)
plt.xlim(left=0, right=0.7)
plt.ylim(bottom=0)
plt.legend(loc='upper right', frameon=False)
plt.xlabel(r'$F/F_{1s}$', fontsize=11)
plt.ylabel(r'$\theta~(°)$', fontsize=12)
plt.show()
```

```python
# print angle values to output file
with open('angle_'+str(g)+'.out', "w") as f:
    for i in range(frac.size-start):
        f.write(str(str(frac[i+start]))+' '+ str(force[i+start])+' '+str( angle[i])+'\n')
f.close()
```
